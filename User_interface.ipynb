```python
# Convert the 'crops' column values to lowercase
sub_crops['crops'] = sub_crops['crops'].str.lower()

# Get the unique labels and unique crops and convert them into sets
unique_labels = set(sorted(crops['label'].unique()))
unique_crops = set(sorted(sub_crops['crops'].unique()))

# Find the common values (intersection)
common_values = unique_labels.intersection(unique_crops)

# Find the different values (symmetric difference)
different_values = unique_labels.symmetric_difference(unique_crops)

# Print the results
print("Common values:", common_values)
print("Different values:", different_values)
```

```python
# Get the unique subdivisions from 'SUBDIVISION' column in 'sub_crops' and sort them
unique_subdivisions = sorted(sub_crops['SUBDIVISION'].unique())

# Create a Dropdown widget for subdivisions
dropdown_subdivisions = widgets.Dropdown(
    options=unique_subdivisions,
    description='Select Subdivision:'
)

# Create a Dropdown widget for crops under the selected subdivision
dropdown_crops = widgets.Dropdown(
    description='Select Crop:'
)

# Create an output widget to display the list of crops
output_crops = widgets.Output()

# Display the dropdown widgets and output widget
display(dropdown_subdivisions, dropdown_crops, output_crops)

# Define a function to update the list of crops based on the selected subdivision
def update_crops_list(change):
    selected_subdivision = change.new
    with output_crops:
        output_crops.clear_output()
        print("Selected Subdivision:", selected_subdivision)
        selected_crops = sub_crops.loc[sub_crops['SUBDIVISION'] == selected_subdivision, 'crops'].unique()
        print("Crops in selected subdivision:")
        for crop in selected_crops:
            print("- ", crop)
        # Update the crops dropdown options
        dropdown_crops.options = selected_crops

# Attach the observe function to the subdivision dropdown widget
dropdown_subdivisions.observe(update_crops_list, names='value')

# Define a function to handle the crop selection
def handle_crop_selection(change):
    selected_subdivision = dropdown_subdivisions.value
    selected_crop = change.new
    with output_crops:
        output_crops.clear_output()
        print(f"Selected Crop: {selected_crop}")

        # Find the 'label' row corresponding to the selected crop
        crop_data = crops[crops['label'] == selected_crop]

        # Calculate the range for each column (N, P, K, temperature, humidity, pH, rainfall)
        min_N = crop_data['N'].min()
        max_N = crop_data['N'].max()
        min_P = crop_data['P'].min()
        max_P = crop_data['P'].max()
        min_K = crop_data['K'].min()
        max_K = crop_data['K'].max()
        min_temp = crop_data['temperature'].min()
        max_temp = crop_data['temperature'].max()
        min_humidity = crop_data['humidity'].min()
        max_humidity = crop_data['humidity'].max()
        min_ph = crop_data['ph'].min()
        max_ph = crop_data['ph'].max()
        min_rainfall = crop_data['rainfall'].min()
        max_rainfall = crop_data['rainfall'].max()

        # Calculate the average for each column (N, P, K, temperature, humidity, pH, rainfall)
        avg_N = crop_data['N'].mean()
        avg_P = crop_data['P'].mean()
        avg_K = crop_data['K'].mean()
        avg_temp = crop_data['temperature'].mean()
        avg_humidity = crop_data['humidity'].mean()
        avg_ph = crop_data['ph'].mean()
        avg_rainfall = crop_data['rainfall'].mean()

        # Display the range and average for each column with two decimals
        print(f"Range for Nitrogen: {min_N:.2f} - {max_N:.2f}, Average: {avg_N:.2f}")
        print(f"Range for Phosphorus: {min_P:.2f} - {max_P:.2f}, Average: {avg_P:.2f}")
        print(f"Range for Potassium: {min_K:.2f} - {max_K:.2f}, Average: {avg_K:.2f}")
        print(f"Range for temperature: {min_temp:.2f} - {max_temp:.2f}, Average: {avg_temp:.2f}")
        print(f"Range for humidity: {min_humidity:.2f} - {max_humidity:.2f}, Average: {avg_humidity:.2f}")
        print(f"Range for pH: {min_ph:.2f} - {max_ph:.2f}, Average: {avg_ph:.2f}")
        print(f"Range for rainfall: {min_rainfall:.2f} - {max_rainfall:.2f}, Average: {avg_rainfall:.2f}")

        # Filter the rainfall data for the selected subdivision and years 2015, 2016, and 2017
        selected_subdivision_rainfall = rainfall[(rainfall['SUBDIVISION'] == selected_subdivision) & (rainfall['YEAR'].isin([2015, 2016, 2017]))]

        # Display the recent rainfall for 2015 and 2016
        recent_rainfall = selected_subdivision_rainfall[selected_subdivision_rainfall['YEAR'].isin([2015, 2016])]
        print("\nRecent Rainfall (Seasonal) for 2015 and 2016:")
        print(recent_rainfall[['Kharif (Avg. Jun-Sept)', 'Rabi (Avg. Oct-Mar)', 'Zaid (Avg. Mar-Jun)', 'ANNUAL']])

        # Display the forecasted rainfall for 2017
        forecasted_rainfall = selected_subdivision_rainfall[selected_subdivision_rainfall['YEAR'] == 2017]
        print("\nForecasted Rainfall (Seasonal) for 2017:")
        print(forecasted_rainfall[['Kharif (Avg. Jun-Sept)', 'Rabi (Avg. Oct-Mar)', 'Zaid (Avg. Mar-Jun)', 'ANNUAL']])

        # Filter the rainfall data for the selected subdivision and year 2017
        selected_subdivision_kharif_rainfall_2017 = rainfall[(rainfall['SUBDIVISION'] == selected_subdivision) & (rainfall['YEAR'] == 2017)]['Kharif (Avg. Jun-Sept)'].iloc[0]

        # Compare the avg rainfall for all crops in the selected subdivision to the 'Kharif (Avg. Jun-Sept)' rainfall for 2017 of that subdivision
        avg_kharif_rainfall_by_crop = sub_crops[sub_crops['SUBDIVISION'] == selected_subdivision].merge(crops, left_on='crops', right_on='label', how='left').groupby('crops')['rainfall'].mean()

        # Calculate the absolute difference between the avg rainfall of each crop and the 'Kharif (Avg. Jun-Sept)' rainfall for 2017
        diff_with_kharif_2017 = avg_kharif_rainfall_by_crop.sub(selected_subdivision_kharif_rainfall_2017).abs()

        # Select the top 3 distinct crops with the closest average Kharif (Avg. Jun-Sept) rainfall to the Kharif (Avg. Jun-Sept) rainfall of the year 2017 for the selected subdivision
        top_3_crops_by_kharif_rainfall = diff_with_kharif_2017.nsmallest(3)

        # Display the top 3 crops by rainfall
        print("\nTop 3 Recommended Crops for Kharif season of 2017 in the Selected Subdivision:")
        for crop, diff_with_kharif_rainfall in top_3_crops_by_kharif_rainfall.items():
            crop_avg_rainfall = crops[crops['label'] == crop]['rainfall'].mean()
            print(f"- Crop: {crop}, Avg Rainfall: {crop_avg_rainfall:.2f}")


# Attach the observe function to the crop dropdown widget
dropdown_crops.observe(handle_crop_selection, names='value')
```
